```python
# Import libararies 
import numpy as np
import matplotlib.pyplot as plot
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
#Sample non_linear data
X = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
y = np.array([1, 4, 9, 16, 25, 36, 64, 81]) #qudratic relationship
```

```python
#Model training
model=LinearRegression()
model.fit(X,y)
```

```python
#Model prediction
y_pred=model.predict(X)
print(y_pred) #predicted value
print(y) #actual value
print(y.mean()) # mean of all the data points
```

```python
#Model evaluation

mse=mean_squared_error(y,y_pred)
print("Mean squared error: \n",mse)

# Compute rmse for the linear model predictions
rmse=np.sqrt(mse)
print(f"Root mean squared error: \n {rmse}")

# Compute mae for the linear model predictions
mae=mean_absolute_error(y,y_pred)
print(f"Mean absolute error: \n {mae}")

# Compute r2-score for the linear model predictions
accuracy=r2_score(y,y_pred)
# It is the amount of the variation in the output dependent attribute which is predictable from the input independent variable
print(f"R2 score: \n {accuracy}")

# Compute z-score for the linear model predictions
z_score = (y_pred - np.mean(y_pred)) / np.std(y_pred) 
# It helps to understand if a data value is greater or smaller than the mean and how far away it is from the mean.
print(f"The Z-Score: \n {z_score}")
```

```python
#Plot 
plot.scatter(X,y,color='b',edgecolors='pink',marker='o',label='Actual fit')
plot.plot(X,y_pred,label='Linear Fit', marker='o')
plot.legend()
plot.grid()
plot.show()
```

```python
#Using polynomial features
poly=PolynomialFeatures(degree=2)
X_poly=poly.fit_transform(X)
print(X_poly)
```

```python
#Model Training
model=LinearRegression()
model.fit(X_poly,y)
```

```python
# Model Prediction
y_pred_poly=model.predict(X_poly)
print(y_pred_poly) 
print(y_pred)
print(y)
```

```python
#Plot
plot.scatter(X,y,color='b',edgecolors='pink',label='Actual fit')
plot.plot(X,y_pred_poly,label='polynomial Fit', marker='o')
plot.legend()
plot.grid()
plot.show()
```

```python
#Model Evaulation
mse=mean_squared_error(y,y_pred_poly)
print(mse)

accuracy=r2_score(y,y_pred_poly)
print(accuracy)
```
